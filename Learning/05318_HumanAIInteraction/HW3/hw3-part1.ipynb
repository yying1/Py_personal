#### Question 7

```python
df_cause = df.loc[df['group'].isin(['Heart Disease', 'Respiratory Condition','Skin Condition','Asthma'])][['age','group']]
x1 = list(df_cause[df_cause['group'] == 'Heart Disease']['age'])
x2 = list(df_cause[df_cause['group'] == 'Respiratory Condition']['age'])
x3 = list(df_cause[df_cause['group'] == 'Skin Condition']['age'])
x4 = list(df_cause[df_cause['group'] == 'Asthma']['age'])

colors=['blue', 'green', 'orange','purple']
names=['Heart Disease','Respiratory Condition','Skin Condition','Asthma']

plt.figure(figsize=(15,15))
plt.hist([x1, x2, x3,x4], color=colors, label=names, density=False)
plt.legend()
plt.title('Side-by-side Histogram for 4 causes of death by age')
plt.xlabel('age')
plt.show() 
```

Side-by-side histogram offers a clear view of different causes of death in each age group. Different color groups are also helpful to distinguish different causes. Instead of having 4 separate charts and different axis scales, this is straightforward to compare.
