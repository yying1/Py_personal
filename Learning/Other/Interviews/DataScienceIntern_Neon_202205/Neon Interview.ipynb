```python
import pandas as pd
import datetime
```

```python
raw_dt = pd.read_csv("Data-Exercise-Transactions.csv")
```

```python
raw_dt
```

```python
raw_dt['date'] = pd.to_datetime(raw_dt['date'], errors ='coerce')
raw_dt['week'] = raw_dt['date'].dt.week
raw_dt['year'] = raw_dt['date'].dt.year
```

```python
dt_2020 = raw_dt[raw_dt["year"]==2020]
```

```python
dt_2020
```

```python
dt1 = dt_2020.groupby(["user_id","week"]).agg({'amount':'sum'})
```
