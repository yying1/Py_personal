```python
# Part a
def read_expenses():
    expense_list = []
    with open("expenses.txt",'r') as f:
        header = f.readline().strip().split(":")
        print("%-15s %-15s %-15s %-15s %-15s"%(header[0],header[1],header[2],header[3],header[4]))
        print(79*"-")
        lines = f.readlines()
        for line in lines: 
            data = line.strip().split(":")
            expense_list.append(data)
            print("%-15s %-15s %-15s %-15s %-15s"%(data[0],data[1],data[2],data[3],data[4]))
    return expense_list
```

```python
# Part b
def read_people():
    people_list = []
    with open("people.txt",'r') as f:
        header = f.readline().strip().split(":")
        print("%-15s %-15s %-15s %-15s"%(header[0],header[1],header[2],header[3]))
        print(63*"-")
        lines = f.readlines()
        for line in lines: 
            data = line.strip().split(":")
            people_list.append(data)
            print("%-15s %-15s %-15s %-15s"%(data[0],data[1],data[2],data[3]))
    return people_list
```

```python
# Part c
def process_peoplelist(a):
    people_list = a
    department_list = [x[3] for x in people_list]
    department_list = list(set(department_list))
    depart_dict = {}
    for d in department_list:
        id_list = [x[0] for x in people_list if x[3] == d]
        depart_dict[d] = id_list
    for d in depart_dict.keys():
        print(20*"-")
        print("Department Name: "+d)
        print("Number of employees: "+str(sum([1 for x in people_list if x[3] == d])))
        print("Employee s: "+";".join([x[2]+" "+x[1] for x in people_list if x[3] == d]))
    return depart_dict
```

```python
# Part d
def expense_report (expense_list,people_list):
    purchase_dict = {}
    purchase_names = []
    purchase_total = []
    for e in expense_list:
        if e[0] in purchase_dict.keys():
            purchase_dict[e[0]].append(e[1:])
        else:
            last_name = [x[1] for x in people_list if x[0].strip() == e[0].strip()][0]
            last_name = last_name+"_"+e[0].strip()
            purchase_names.append(last_name)
            value_list = [e[1:],]
            purchase_dict[e[0]] = value_list
    print(f'{"Name":20s}\t{"ID":15s}\t{"Amount":10s}\t{"Dates":15s}\t{"Description":15s}')
    print(f'{"----":25s}\t{"--":15s}\t{"------":15s}\t{"-----":15s}\t{"-----------":15s}')
    for n in list(sorted(purchase_names)):
        total_expense = 0
        e_id = n.split("_")[-1]
        name= n.split("_")[0]
        for exp in purchase_dict[e_id]:
          amount = exp[0]
          date = str(exp[2])
          date = "/".join((date[4:6],date[-2:],date[0:4]))
          if total_expense == 0:
              print(f'{name:20s}\t{e_id:15s}\t{amount:15s}\t{date:9s}\t{exp[3]:15s}')
          else:
              print(f'{"    ":25s}\t{"  ":20s}\t{amount:15s}\t{date:9s}\t{exp[3]:15s}')
          total_expense += float(amount)
        print(f'{"    ":25s}\t{"  ":20s}\t{"------":10s}\t{"     ":15s}\t{"           ":15s}')
        print(f'{"    ":25s}\t{"  ":20s}\t{total_expense:.2f}\t{"     ":15s}\t{"           ":15s}')
        purchase_total.append([name,e_id,total_expense])
    return purchase_total
```

```python
# Part e
def generate_smmary(purchase_total):
    with open("summary.txt","w") as f:
        f.write("lastName,id,totalAmount"+"\n")
        for i in purchase_total:
            purchase_list = [str(x) for x in i]
            f.write(",".join(purchase_list)+"\n")
```

```python
# Part f
def add_newrecord ():
    with open("expenses.txt","a") as f:
        f.write("5:22.95:supply:20190412:Bic pens"+"\n")
        f.write("4:149.95:travel:20190415:Car rental"+"\n")
        f.write("18:77.75:meal:20190416:Lunch with investors"+"\n")
        f.write("7:950.15:travel:20190418:flight to Chicago"+"\n")
```

```python
def main():
    import sys
    print("Part a:")
    expense_list = read_expenses()
    print("Part b:")
    people_list = read_people()
    print("Part c:")
    depart_dict = process_peoplelist(people_list)
    print("Part d:")
    purchase_total = expense_report (expense_list,people_list)
    generate_smmary(purchase_total)
    add_newrecord ()
    print("Part g:")
    expense_list = read_expenses()
```
