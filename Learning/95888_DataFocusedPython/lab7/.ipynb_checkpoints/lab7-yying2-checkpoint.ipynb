```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
M = np.genfromtxt("weather.csv",delimiter=',',dtype = int)
print(M)
```

```python
rows = M.shape[0]
plt.plot(range(rows),M[:,1],label="Temperature",color='red')
plt.title ('Temperature')
plt.legend(loc="upper left")
plt.show()
```

```python
plt.title ('Humidity')
plt.scatter(range(rows),M[:,2],label = 'Humidity',)
plt.legend(loc="upper right")
plt.show()
```

```python
plt.title ('Wind')
plt.bar(range(rows),M[:,3],label = 'Wind',)
plt.legend(loc="upper right")
plt.show()
```

```python
plt.plot(range(rows),M[:,1],label="Temperature",color='red')
plt.title ('Temperature with ylim')
plt.legend(loc="upper left")
plt.ylim(0, max(M[:,1]))
plt.show()
```

```python
data = pd.read_csv("popAreaHousing.csv")
print(data)
```

```python
print(data.corr())
x = round(data.corr().loc['Population']['Area'],2)
y = round( data.corr().loc['Population']['Housing Units'],2)
z = round( data.corr().loc['Area']['Housing Units'],2)
print("Population-Area: ",x,", Population-Housing Units: ",y,", Area-Housing Units: ",z)
```

```python
plt.title ('Pop X Area')
plt.xlabel('Population')
plt.ylabel('Area')
plt.scatter(data['Population'],data['Area'],color ="blue" ,marker = "*")
plt.annotate("Correlation = "+str(x), xy = (0.5,0.1),xycoords = 'axes fraction')
plt.show()
```

```python
plt.title ('Pop X Housing Units')
plt.xlabel('Population')
plt.ylabel('Housing Units')
plt.scatter(data['Population'],data['Housing Units'],color ="darkgreen" ,marker = "*")
plt.annotate("Correlation = "+str(y), xy = (0.5,0.1),xycoords = 'axes fraction')
plt.show()
```

```python
plt.title ('Area X Housing Units')
plt.xlabel('Population')
plt.ylabel('Housing Units')
plt.scatter(data['Population'],data['Housing Units'],color ="red" ,marker = "*")
plt.annotate("Correlation = "+str(z), xy = (0.5,0.1),xycoords = 'axes fraction')
plt.show()
```

```python
plt.title ('Population Pie Chart')
label = list(data['State'])
plt.pie(data['Population'],labels = label)
plt.show()
```

```python
plt.title ('State Versus Population')
plt.bar(data['State'],data['Population'])
plt.xticks(rotation = 45)
plt.xlabel('State')
plt.show()
plt.title ('State Versus Population with ti')
plt.bar(data['State'],data['Population'])
plt.xticks([])
plt.xlabel('State')
plt.show()
```

```python
sns.pairplot(data = data)
```

```python
def f(row):
    if row['Population']>= 10000000:
        val = 'Big'
    elif 10000000> row['Population']>= 5000000:
        val = 'Medium'
    else:
        val = 'Small'
    return val
data['Size'] = data.apply(f,axis = 1)
```

```python
snsdata2 = data[["Size","Population"]]
sns.boxplot(x = "Size", y ='Population', data = snsdata2)
```
