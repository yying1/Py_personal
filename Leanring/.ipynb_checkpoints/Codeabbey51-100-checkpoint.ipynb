```python
# 2021-08-19 Problem 51
# Time: 56 mins
## from random import randrange ## generate random number
import numpy as np
d_num = [1,2,3,4,5]
d_sides = [2,4,6,8,10,12]
d_comb = []
for n in d_num:
    for s in d_sides:
        c_min = n*1
        c_max = n*s
        c_avg = (1+s)/2*n
        c_dic = {"min":c_min,"max":c_max,"avg":c_avg,"answer":str(n)+'d'+str(s)}
        d_comb.append(c_dic)

file = open("51.txt")
result_list=[]
for l in range(3):
    raw_list = file.readline().split(" ")
    del raw_list[-1]
    num_list = [int(x) for x in raw_list]
    d_comb_f = [x for x in d_comb if x["min"] <= min(num_list) and x["max"] >= max(num_list)]
    if len(d_comb_f) == 1:
        result = d_comb_f[0]["answer"]
    else:
        dis = abs(d_comb_f[0]["avg"]-np.mean(num_list))
        result = d_comb_f[0]["answer"]
        for i in d_comb_f[1:]:
            if abs(i["avg"]-np.mean(num_list)) < dis:
                result = i["answer"]
    result_list.append(result)
print(*result_list," ")
file.close()
```

```python
# 2021-08-23 Problem 53
# Time: 90 mins
import string 
import itertools

f = open("53.txt")
num = int(f.readline())
result_list = []

def gen_eaten(q,board_size):
    eaten_list = []
    ranks = list(range(1,int(board_size)+1))
    files = list(string.ascii_lowercase)[0:int(board_size)]
    board = list(itertools.product(files, ranks))
    index_diff = ranks.index(int(q[1:2])) - files.index(q[0:1])
    index_sum = ranks.index(int(q[1:2])) + files.index(q[0:1])
    #print(board)
    for p in board:
        if p[0] == q[0:1] or p[1] == int(q[1:2]) or ranks.index(p[1]) - files.index(p[0])  == index_diff or ranks.index(p[1]) + files.index(p[0]) == index_sum:
            eaten_list.append(p) 
            #print(p)
    return eaten_list
    

for n in range(num):
    k,q = f.readline().split(" ")
    eaten_list = gen_eaten(q,8)
    if (k[0:1],int(k[1:2])) in eaten_list:
        result = "Y"
    else:
        result = "N"
    result_list.append(result)
        
print(*result_list," ")
f.close()
```

```python
## 2021-08-26 Problem #56
# Time: 125 mins
import itertools

def enlarge_a(a,r,c):
    a1 = np.zeros([r,c],dtype = int)
    for idx, x in np.ndenumerate(a):
        a1[idx[0]+1][idx[1]+1] = x
    return a1
    
def count_a(a,r,c):
    ri,ci = 0,0
    b = np.zeros([r,c],dtype = int)
    #print(ri,ci,r,c)
    while ri < r:
        ci=0
        while ci<c:
            ri_range = [max(0,ri-1),ri,min(r-1,ri+1)]
            ci_range = [max(0,ci-1),ci,min(c-1,ci+1)]
            neighbors = list(set(list(itertools.product(ri_range,ci_range))))
            count = 0
            for n in neighbors:
                if n[0] == ri and n[1] ==ci:
                    continue 
                if a[n[0]][n[1]] == 9:
                    count+=1
            #print(count,ri,ci)
            b[ri][ci] = count
            ci+=1
        ri+=1
    return b

def gen_new(a,b,r,c):
    d = np.zeros([r,c],dtype = int)
    for idx, x in np.ndenumerate(a):
        #print(idx)
        if x != 9:
            if b[idx[0]][idx[1]] == 3:
                d[idx[0]][idx[1]] = 9
        else:
            if b[idx[0]][idx[1]] in (2,3):
                d[idx[0]][idx[1]] = 9 
            
    return d
            
import numpy as np
a = np.zeros([5,7],dtype = int)
f = open("56.txt")
result_list = []
for r in range (5):
    row = list(f.readline().strip())
    for ci,c in enumerate(row):
        if c == 'X':
            a[r][ci] = 9
r,c = 5,7
for i in range(5):
    a = enlarge_a(a,r+2,c+2)
    b = count_a(a,r+2,c+2)
    d = gen_new(a,b,r+2,c+2)
    count = np.count_nonzero(d == 9)
    result_list.append(count)
#     print(a)
#     print(b)
#     print(d)
#     print(count," next", i+1)
    a = d.copy()
    r+=2
    c+=2
print(*result_list," ")
f.close()
```
