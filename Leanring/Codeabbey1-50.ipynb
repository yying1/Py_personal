```python
## 2021-07-1 codeabbey problem ID:19
# took about 5 hours
file = open("19.txt")
number = file.readline()
result_list = []
pair_dic = {}
test_list = {"{":"}","(":")","[":"]","<":">"}
compare_list = {"{":"( [ <","(":"{ [ <","[":"{ ( <","<":"{ ( ["}
for i in range(int(number)):
    raw_list = list(file.readline())
    result = True
    for begin in ["(","[","{","<"]:
        close = test_list[begin]
        position_list = []
        pair_list = []
        stck = []
        for index, value in enumerate(raw_list):
            if value in (begin,close):
                group = (value,index)
                position_list.append(group)
        if len(position_list) % 2 != 0:
            result = False
            break
        try:
            for ele1, ele2 in position_list:
                if begin in ele1:
                    stck.append((ele1, ele2))
                elif close in ele1:
                    pair_list.append((stck.pop()[1], ele2))
            pair_dic[begin] = pair_list
        except:
            result = False
            break
    if result == False:
        result_list.append(0)
        continue
    for i in ["(","[","{","<"]:
        base_groups = pair_dic[i]
        for group in base_groups:
            start = group[0]
            end = group[1]
            for compare in compare_list[i].split(" "):
                compare_groups = pair_dic[compare]
                for c_group in compare_groups:
                    if start < c_group[0] < end and end < c_group[1]:
                        result = False
                        break
                    if c_group[0]< start and start < c_group[1] <end:
                        result = False
                        break
            if result == False:
                break
    if result == False:
        result_list.append(0)
    else:
        result_list.append(1)
print(*result_list, sep = " ")
file.close()
```

```python
## 2021-07-1 codeabbey problem ID:19
# advanced version
file = open("19.txt")
number = file.readline()
result_list = []
start_list = ["(","[","{","<"]
end_list = [")","]","}",">"]
pair_dict = {"}":"{",")":"(","]":"[",">":"<"}
for i in range(int(number)):
    raw_list = list(file.readline())
    brackets = []
    continue_1 = True
    for ii in raw_list:
        if ii in start_list:
            brackets.append(ii)
        if ii in end_list:
            try:
                if pair_dict[ii] == brackets[-1]:
                    del brackets[-1]
                else:
                    result_list.append(0)
                    continue_1 = False
                    break
            except:
                result_list.append(0)
                continue_1 = False
                break
    if continue_1 != False and len(brackets) > 0:
        result_list.append(0)
    elif continue_1 == True and len(brackets) == 0:
        result_list.append(1)
print(*result_list, sep = " ")
```

```python
## 2021-07-08 codeabbey problem ID:26
# 20 mins
from functools import reduce
file = open("26.txt")
number = file.readline()
result_list = []
for i in range(int(number)):
    raw_list = file.readline().split(" ")
    raw_list = [int(x) for x in raw_list]
    if min(raw_list) == 0:
        GCD = max(raw_list)
    else:
        a = min(raw_list)
        b = max(raw_list)
        while True:
            if b % a == 0:
                GCD = a
                break
            else:
                a,b = b % a,a
    lcm = int(reduce(lambda x,y: x*y,raw_list,1)/GCD)
    result_list.append(str(GCD)+" "+str(lcm))
result_string = ") (".join(result_list)
print("("+result_string+")")
```

```python
## 2021-07-15 codeabbey problem ID:34
# Time: 75 min
import math as m

def get_answer(x,a,b,c,d):
    answer = a*x + b*m.sqrt(x**3.0) - c*m.exp(-x/50) - d
    return answer

file = open("34.txt")
number = file.readline()
result_list = []
for i in range(int(number)):
    raw_list = file.readline().split(" ")
    raw_list = [float(x) for x in raw_list]
    a,b,c,d = raw_list[0],raw_list[1],raw_list[2],raw_list[3]
    x1,x2 = 0,100
    while True:
        xm = (x1+x2)/2
        answer_dic = {x1:0,x2:0,xm:0}
        for i in [x1,x2,xm]:
            answer_dic[i] = get_answer(i,a,b,c,d)
        #print(str(x1)+" "+str(x2)+" "+str(abs(x1-x2)))
        if answer_dic[x1] < 0 < answer_dic[xm]:
            x1,x2 = x1,xm
        else:
            x1,x2 = xm,x2
        if abs(x1-x2) < 0.00000001:
            result_list.append((x1+x2)/2)
            break
print(*result_list," ")
file.close()
```

```python
# 2021-07-28 problem 42
# Time: 45 mins
file = open("42.txt")
number = int(file.readline())

point_counter = {"T":10,"J":10,"Q":10,"K":10}
final_result= []

def count_aces(num):
    result = [1,11]
    point_aces = [1,11]
    while num > 1:
        temp = []
        for p in result:
            for i in point_aces:
                sumpi = p+i
                if sumpi not in temp:
                    temp.append(sumpi)
        result = temp
        num -=1
    return result
        
def sum_cards(cards):
    sum1 = 0
    result = []
    for c in cards:
        if c in point_counter.keys():
            sum1 += point_counter[c]
        if c not in point_counter.keys() and c != "A":
            sum1 += int(c)
    aces_count = sum(1 for x in cards if x == "A")
    if aces_count > 1:
        result = count_aces(aces_count)
    elif aces_count ==1:
        result = [1,11]
    else:
        result = [0]
    result = [x + sum1 for x in result]
    return result

for i in range(number): 
    raw_list = file.readline().strip().split(" ")
    result = sum_cards(raw_list)
    if min(result) > 21:
        final_result.append("Bust")
    else:
        value = 0
        for i in result:
            if i > value and i <= 21:
                value = i
        final_result.append(str(value))
        
print(*final_result," ")
```

```python
# 2021-08-03 problem 46
# Time = 40 mins
result_list = []
win_sets = [[0,1,2],[3,4,5],[6,7,8],[0,3,6],[1,4,7],[2,5,8],[0,4,8],[2,4,6]]

def check_win(board,win_sets):
    result = False
    for ws in win_sets:
        if board[ws[0]] == board[ws[1]] and board[ws[0]] == board[ws[2]]:
            result = True
            break
    return result

file = open("46.txt")
number = int(file.readline())
for i in range(number):
    board = [1,2,3,4,5,6,7,8,9]
    game_steps = file.readline().strip().split(" ")
    game_steps = [int(x) for x in game_steps]
    result = 0
    for i,v in enumerate(game_steps):
        if i%2 == 0:
            board[v-1] = "x"
        else:
            board[v-1] = 0
        status = check_win(board,win_sets)
        #print(board,status)
        if status == True:
            result = i+1
            break
    result_list.append(result)
print(*result_list," ")
file.close()
```
