```python
#2021-11-11 Problem #67
#Time: 42 mins
def rec_f_withlist(a,b_list):
    dic = {0:0,1:1,2:1}
    s = 1
    n = 1
    count = 2
    while n != a:
        n = s + n
        s = n - s
        count +=1
        if n in b_list:
            dic[n] = count
    dic[a] = count
    return dic

with open("67.txt","r") as f:
    cases = int(f.readline())
    f_list = list()
    for c in range(cases):
        f_list.append(int(f.readline()))
f_slist = sorted(f_list)
result = rec_f_withlist(f_slist[-1],f_slist[:-1])
for i in f_list:
    print(result[i],end=" ")
```

```python
# 2021-12-28 Problem #73 Hexagonal Grid
# 17 mins
def move(start,point):
    if point == "A":
        start[0] = start[0]+1
    elif point == "B":
        start[0] = start[0]+1/2
        start[1] = start[1]+(3/4)**0.5
    elif point == "C":
        start[0] = start[0]-1/2
        start[1] = start[1]+(3/4)**0.5
    elif point == "D":
        start[0] = start[0]-1
    elif point == "E":
        start[0] = start[0]-1/2
        start[1] = start[1]-(3/4)**0.5
    elif point == "F":
        start[0] = start[0]+1/2
        start[1] = start[1]-(3/4)**0.5
    return start

file = open("73.txt")
cases = int(file.readline())
result_list = []
for c in range(cases):
    moves = list(file.readline())
    start = [0.0,0.0]
    for m in moves:
        start = move(start,m)
    distance = round(((start[0]-0.0)**2 + (start[1]-0.0)**2)**0.5,8)
    result_list.append(distance)
print(*result_list," ")    
```
