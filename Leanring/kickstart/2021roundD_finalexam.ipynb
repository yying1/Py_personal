```python
## 2021-08-31
# Round D: Final Exam, with test 2 dataset
# Note: test 2 answer seems to be right, but somehow it is still taking longer time to process

import numpy as np
from collections import OrderedDict

def match_pe_ps(pe_key,pe_val,ps_key,ps_val,s):
    pos = ""
    if pe_val<= s <= ps_key:
        if abs(s-pe_val) > abs(s-ps_key):
            match_s = ps_key
            match_e = ps_val
            if ps_key<s<ps_val:
                pos = "M"
            elif ps_key >= s:
                pos = "K"
        else:
            match_s = pe_key
            match_e = pe_val
            if pe_key<s<pe_val:
                pos = "M"
            elif pe_val <= s:
                pos = "V"
    elif s<pe_val:
        match_s = pe_key
        match_e = pe_val
        if pe_key>=s:
            pos = "K"
        else:
            pos = "M"
    elif ps_key<s:
        match_s = ps_key
        match_e = ps_val
        if ps_val<= s:
            pos = "V"
        else:
            pos = "M"
    return match_s,match_e,pos

f = open("ts2_input.txt")
# ff = open("1.txt","a")
cases = int(f.readline())

# cases = int(input())

for case in range(cases):
    p_s,students = f.readline().split(" ")
#     p_s,students = input().split(" ")
    problems = {}
    assignments = []
    for s in range(int(p_s)):
        b,e = f.readline().split(" ")
#         b,e = input().split(" ")
        problems[int(b)] = int(e)
        # Creates a sorted dictionary (sorted by key)
#     problems = OrderedDict(sorted(problems.items()))
#         print(problems)

    student_skills = f.readline().split(" ")
#     student_skills = input().split(" ")
    student_skills = [int(x) for x in student_skills]
#     print(len(problems),len(student_skills))
    keyList=list(sorted(problems.keys()))
    for s in student_skills:
#         print(s)
        keyval = min(problems.items(), key=lambda x: abs(s - x[0]))
        valval = min(problems.items(), key=lambda x: abs(s - x[1]))
#         if s == 840:
#             print(keyval,valval)
        if keyval == valval:
            keyval =abs(keyval[0] - s)
            valval =abs(s-valval[1])
            ps_key_l = [k for k, v in problems.items() if k-s in (keyval,0-keyval)]
            pe_key_l = [k for k, v in problems.items() if s-v in(valval,0-valval)]
            #print(type(ps_key_l),type(pe_key_l),len(ps_key_l),len(pe_key_l))
#             if len(ps_key_l)>0:
            ps_key,ps_val = min(ps_key_l),problems[min(ps_key_l)]
#             else:
#                 ps_key,ps_val = min(pe_key_l),problems[min(pe_key_l)]
#             if len(pe_key_l)>0:
            pe_key,pe_val = min(pe_key_l),problems[min(pe_key_l)]
#             else:
#                 pe_key,pe_val = min(ps_key_l),problems[min(ps_key_l)]
        else: 
            pe_key, pe_val = min(problems.items(), key=lambda x: abs(x[1]-s))
            ps_key, ps_val = min(problems.items(), key=lambda x: abs(s-x[0]))
#         if s == 840:
#             print(pe_key,pe_val,ps_key,ps_val)
        closest = 0
        match_s, match_e, pos = match_pe_ps(pe_key,pe_val,ps_key,ps_val,s)
        #print(match_s, match_e, pos)
        if pos == "K":
            del problems[match_s]
            if match_s < match_e:
                problems[match_s+1] = match_e
            closest = match_s
        elif pos == "V":
            if match_s < match_e:
                problems[match_s] = match_e-1
            else:
                del problems[match_s]
            closest = match_e
        elif pos == "M":
            if match_s <= s-1:
                problems[match_s] = s-1
            if s+1<= match_e:
                problems[s+1] = match_e
            closest = s
        assignments.append(closest)
    print(f'Case #{str(case+1)}: {" ".join(str(x) for x in assignments)}')
```
