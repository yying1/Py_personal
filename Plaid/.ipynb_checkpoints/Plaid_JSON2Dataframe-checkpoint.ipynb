```python
import json
import pandas as pd
```

```python
with open('asset_report.json') as json_file:
    data = json.load(json_file)
```

```python
data.keys()
```

```python
 data['items'][0]['accounts']
```

```python
# items contains all acounts, list accout names
for account in data['items'][0]['accounts']:
    if account['official_name'] is None:
        print(account['name'])
    else:
        print(account['official_name'])
print(data['items'][0]['accounts'][0].keys())
```

```python
# Get checking account transactions, convert to data frames
transaction_headers = list(data['items'][0]['accounts'][0]['transactions'][0].keys())
df = pd.DataFrame (data['items'][0]['accounts'][0]['transactions'], columns =transaction_headers)
df
```
