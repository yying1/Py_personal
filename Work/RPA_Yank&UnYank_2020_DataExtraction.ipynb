```python
import pandas as pd
```

```python
df = pd.read_csv('SIM_Process_Log.txt', header = None)
```

```python
df['Robot'] = df[0]
df['Time'] = df[0]
df['SIM'] = df[0]
df['SIM_Count']=''
```

```python
df
```

```python
for index, row in df.iterrows(): 
    row['Robot'] = row[0].strip().split('-')[0].strip()
    row['Time'] = row[0].strip().split('-')[1].strip()
    row['SIM'] = row[0].strip().split('-')[2].strip()
    row['SIM_Count'] = row[0].count(';')
```

```python
df.to_excel('RPA_Yank&UnYank_2020_DataExtraction.xlsx', index = False)
```
