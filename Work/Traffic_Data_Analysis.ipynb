```python
import pandas as pd
```

```python
OCTH = pd.read_excel(r"C:\Users\yingyy\Desktop\Traffic_Analysis_20210504.xlsx",sheet_name = 'OCTH_ASIN_202104')
OTHAD = pd.read_excel(r"C:\Users\yingyy\Desktop\Traffic_Analysis_20210504.xlsx",sheet_name = 'OTHAD_ASIN_202104')
OTHRAD = pd.read_excel(r"C:\Users\yingyy\Desktop\Traffic_Analysis_20210504.xlsx",sheet_name = 'OTHRAD_ASIN_202104')
```

```python
OCTH = OCTH.rename(columns=lambda x: 'OCTH_'+x)
```

```python
df_join = pd.merge(OCTH,OTHAD,left_on=['OCTH_session_id','OCTH_request_id'],right_on =['OTHAD_session_id','OTHAD_request_id'],how='outer')
```

```python
df_join = pd.merge(df_join,OTHRAD,left_on=['OCTH_session_id','OCTH_request_id'],right_on =['OTHRAD_session_id','OTHRAD_request_id'],how='outer')
```

```python
df_join.to_csv(r"C:\Users\yingyy\Desktop\Result.csv", index = False, encoding = 'UTF-8')
```
